```python
"""
Setting up Catalog, Schemas and Volumes for
CMAPPS Predictive Maintenance Pipeline
catalog: cmapss_project
├── schema: source_raw   <-- raw .txt files 
├── schema: bronze       <-- Delta tables, raw ingested
├── schema: silver        <-- cleaned, unified, feature-engineered
└── schema: gold           <-- aggregated analytics tables
"""
```

```python
%sql
CREATE CATALOG IF NOT EXISTS cmapss_project
COMMENT 'End-to-end predictive maintenance pipeline using NASA CMAPSS'
```

```python
%sql
USE CATALOG cmapss_project;
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS cmapss_project.source_raw
COMMENT 'Raw source files uploaded from NASA CMAPSS dataset - .txt files';
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS cmapss_project.bronze
COMMENT 'BRONZE-layer - raw ingested Delta tables with metadata'
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS cmapss_project.silver
COMMENT 'SILVER-layer - cleaned, unified, RUL computed, ML-ready'
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS cmapss_project.gold
COMMENT 'GOLD-layer - aggregated analytics table for BI and dashboards'
```

```python
%sql
CREATE VOLUME IF NOT EXISTS cmapss_project.source_raw.raw_files
COMMENT 'Landing zone for raw NASA CMAPSS .txt files';
```
